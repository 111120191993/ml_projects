```python
# Basic Deep Learning Package
import tensorflow as tf
from tensorflow import keras
```

```python
# import the Boston Dataset (this data is dedicated for regression problem)
from sklearn.datasets import load_boston

# dataset - https://www.cs.toronto.edu/~delve/data/boston/bostonDetail.html
```

```python
X, y = load_boston(return_X_y= True)
```

```python
print(X.shape)
print(y.shape)
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state = 101)

print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

# Deep Learning Concept - Neural Networks
## Multilayer Perceptron
## API - sequential or Functional

```python
# Sequential API Setup

DNN = tf.keras.models.Sequential()

# Input Layer connected with First Hidden Layer

DNN.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))


# Input Layer connected with Second Hidden Layer
DNN.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

# Adding output layer

DNN.add(tf.keras.layers.Dense(units=1))
```

```python
# Training Deep Learning Model for Regression Problem

DNN.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Predict the model
DNN.fit(X_train, y_train, batch_size = 32, epochs= 200)
```

```python
DNN.fit(X_train, y_train, batch_size = 32, epochs= 200)
```

```python
# Evaluate the model with test dataset

y_pred = DNN.predict(X_test)
print(y_test)  # actual 
print("#############This will separate the actual and predicted line")
print(y_pred) # predicted
```
