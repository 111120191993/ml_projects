```python
def viewImage(image, name_of_window):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window,image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
(h,w,d) = img.shape
center = (w//2,h//2)
M = cv2.getRotationMatrix2D(center, 150,1.0)
rotated = cv2.warpAffine(img, M, (w,h))
viewImage(rotated, "Image after rotation by 180 degree")
```

```python
color_list = [method for method in dir(cv2) if method.startswith('COLOR_') is True]
color_list
len(color_list)
img = cv2.imread("puppy.PNG")
for coli in range(len(color_list)):
    col_name = color_list[coli]
    colr_code = "cv2."+ col_name
    resized = cv2.resize(img,(400,400))
    try:
        rgb_image = cv2.cvtColor(resized,eval(colr_code))
        cv2.imshow(col_name+"_"+str(coli),rgb_image)
        cv2.waitKey()
        
    except Exception as e:
        print("Error on ",str(colr_code))
        pass
    cv2.destroyAllWindows()
```

# Face Detector

```python
import numpy as np
import cv2
cap = cv2.VideoCapture(0)
cap.set(3,640) # set width
cap.set(4,480) # set height
while(True) :
    ret, frame = cap.read()
    frame = cv2.flip(frame,-1) # flip camera verically/reverse order
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('colorful frame', frame)
    cv2.imshow('black and white', gray)
    
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

# Collect images for face recognition

```python
# Load HaarCascade xml file
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
def face_extractor(img):
    faces = face_classifier.detectMultiScale(img, 1.3,5)    
    # face not found
    if faces is ():
        return None    
    # if face found
    for (x, y, w, h) in faces:
        x = x -10
        y = y -10
        cropped_face = img[y: y+h+50, x:x+w+50]
    return cropped_face
cap = cv2.VideoCapture(0) # enable the camera
count = 0
while True:
    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count +=1
        face = cv2.resize(face_extractor(frame),(480,600))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)        
        file = './Images/' + str(count) + '.jpg'
        cv2.imwrite(file, face)        
        cv2.putText(face, str(count), (50,50), cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
        cv2.imshow("Face Cropper", face)
    else :
        print("Face not found")
        pass
    if cv2.waitKey(1)== 13 or count ==100:
        break
cap.release()
cv2.destroyAllWindows()
print("We have successfully collected 100 samples of images")        
```
